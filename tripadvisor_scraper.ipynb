```python
def get_HTML_item_from_urls(urls, chromeDriver):
    """
    For each url, look for the review card.
    For each review card. Look for the following attributes:
    - The rating
    - The card header
    - The card body
    - Optional: The date of review
    - Optional: A Boolean Indicator for Mobile Users
    """
    # chromeDriver.get(urls[0])
    # items = chromeDriver.find_elements(By.CLASS_NAME, "label.container")
    # print(items)
    # item = items[0].find_elements(By.TAG_NAME, "input")
    # print(item)
    # #items = driver.find_element_by_xpath(f'//input[@name="{name}"]')
    # item[0].click()
    # time.sleep(2)
    i = 0
    res = []
    for url in urls:
        chromeDriver.get(url)
        items = chromeDriver.find_elements(By.CLASS_NAME, "label.container")
        print(items)
        item = items[0].find_elements(By.TAG_NAME, "input")   
        item[0].click()
        time.sleep(2)
        
        # if i == 0:
        #     items = maindriver[0].find_elements(By.CLASS_NAME, "label.container")
        #     print(items)
        #     item = items[0].find_elements(By.TAG_NAME, "input")
        #     # print(item[0].click())
        #     #items = driver.find_element_by_xpath(f'//input[@name="{name}"]')
        #     item[0].click()
        #     time.sleep(1)
            
            
        print(i)
        i = i + 1
                    
        for element in chromeDriver.find_elements(By.CLASS_NAME, "review-container"):
            obj = {}
            
            # Find Review Number
            if element.find_elements(By.CLASS_NAME, "ui_bubble_rating.bubble_50"):
                num = 5
            elif element.find_elements(By.CLASS_NAME, "ui_bubble_rating.bubble_40"):
                num = 4
            elif element.find_elements(By.CLASS_NAME, "ui_bubble_rating.bubble_30"):
                num = 3
            elif element.find_elements(By.CLASS_NAME, "ui_bubble_rating.bubble_20"):
                num = 2
            elif element.find_elements(By.CLASS_NAME, "ui_bubble_rating.bubble_10"):
                num = 1
            else:
                num = 0
            
            # print(review[0])
            # # If the review number can't be found, then move on. This is not the tag you are looking for
            # if review == None:
            #     continue
            # # Get Review number using regex
            # num = int(
            #     re.findall(
            #         r'(?<=<span class="ui_bubble_rating.bubble_)(\d+).*(?=>)',
            #         str(review[0]),
            #     )[0]
            # )
            # # Find Card Header
            # card_head = element.find(attrs={"class": "noQuotes"})
            # Find Card username
            card_user = element.find_elements(By.CLASS_NAME, "info_text")
            user_name = card_user[0].find_elements(By.TAG_NAME,'div')[0]
            # Find Card Body ('div', attrs={'class': 'foo'})
            card_body = element.find_elements(By.CLASS_NAME, "partial_entry")[0]
            # Find Date of Review
            review_date_raw_string = element.find_elements(
                By.CLASS_NAME, "prw_rup.prw_reviews_stay_date_hsx"
            )[0]
            review_date_string = re.findall(
                r"(?<=Date of visit: )(.*)", review_date_raw_string.text
            )[0]
            review_date = datetime.strptime(review_date_string, "%B %Y").date()

            obj.update(
                name=user_name.text,
                rating=num,
                review=card_body.text,
                language=detect(card_body.text),
                review_date=review_date,
                platform='tripadvisor'
            )
            res.append(obj)
    return res
```
