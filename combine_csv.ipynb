```python
import glob
 
# list all csv files only
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
```

```python
import pandas as pd
 
l = []
 
for f in csv_files:
    l.append(pd.read_csv(f))
     
df_res = pd.concat(l, ignore_index=True)
df_res

```

```python
df_res.to_csv('output/feedback.csv', index=False)
```
